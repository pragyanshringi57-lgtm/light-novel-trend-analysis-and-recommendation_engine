```python
import pandas as pd
from data_code import collect_light_novels

# Load the raw data from data_code.py
data = collect_light_novels(max_pages=5)

# checking for missing values
data.isnull().sum()
# filling the missing values those with numeric type with the mean the mean of the column
# and those with non numeric type with "unknown"
col=data.isna().sum()>0
tru_col=col[col==True]
print(tru_col)
for i in tru_col.keys():
    if pd.api.types.is_numeric_dtype(data[i]):# Only calculate the mean if the column contains numbers (
        data[i] = data[i].fillna(data[i].mean()) # Fill missing values with the mean of the column
    else:
        data[i]=data[i].fillna("unknown")
data.isna().sum()
```

```python
# looking for duplicates and dropping them
dup=data.duplicated()
data[dup]
print(f"Number of duplicate rows: {dup.sum()}")
data.drop_duplicates(inplace=True)
print(f"Number of duplicate rows after dropping: {data.duplicated().sum()}")
```

```python
# converting all text columns to lowercase for easier analysis and avoid the problem of case sensitivity
#  for example, "Action" and "action" would be treated as the same genre after this transformation, which is important for accurate analysis and grouping of data.
data.info()
data['Title']=data['Title'].astype(str).str.lower()# the reason to use .str is due to panda column is a series and and not aa string so we use .str to acsees the string methods available in pandas series.
data['Synopsis']=data['Synopsis'].astype(str).str.lower()
data['Genres']=data['Genres'].astype(str).str.lower()
data['Themes']=data['Themes'].astype(str).str.lower()   
data['Demographics']=data['Demographics'].astype(str).str.lower()
```

```python
# catorizing the genres into a list of genres for each novel
data["Genres"] = data["Genres"].str.split(", ")
# using hot-encofing to convert the list of genre into a binary matrix
data_genres=data["Genres"].str.get_dummies()
print(data_genres.head())
```
